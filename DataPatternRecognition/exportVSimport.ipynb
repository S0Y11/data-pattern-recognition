```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import mpl_toolkits
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
cities = {
    'city' : ["somalia","United Arab Emirates","Saudi Arabia","Oman","Djibouti","India","Macau","Qatar","Japan","Bahrain", "China"],
    'lat'  :[5.1521,23.4241,23.8859, 21.4735, 11.8251, 20.5937,22.1987,25.3548, 36.2048,26.0667,35.8617],
    'lon'  :[46.1996,53.8478,45.0792, 55.9754, 42.5903, 78.9629,113.5439,51.1839,138.2529,50.5577,104.1954]
}
df = pd.DataFrame(cities, columns = ['city','lon','lat'])
```

```python
m = Basemap (llcrnrlon=30,
            llcrnrlat=-15,
            urcrnrlon=120,
            urcrnrlat=45,
            projection='merc')
m.drawmapboundary(fill_color='white', linewidth=0)
m.fillcontinents(color='#d1cdbf', alpha = 0.7)
m.drawcoastlines(linewidth=0.1, color="white")

for startIndex, startRow in df.iterrows():
    for endIndex in range(startIndex, len(df.index)):
        endRow = df.iloc[endIndex]

        if startRow.city == endRow.city:
            pass
        else:
            m.drawgreatcircle (startRow.lon, startRow.lat, endRow.lon, endRow.lat,
                              linewidth=1, color= '#647add')
            
for i, row in df.iterrows():
    plt.annotate(row.city, xy=m(row.lon+3, row.lat), verticalalignment='center')            
```
