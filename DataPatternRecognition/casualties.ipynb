```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Nonstate_v24_1.csv")
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip()
somalia_df = df[df['location'] == 'Somalia']
```

```python
print(somalia_df)
```

```python
somalia_df['start_date2'] = pd.to_datetime(somalia_df['start_date2'], errors='coerce')
```

```python
somalia_df = somalia_df.sort_values('start_date2')
```

```python
plt.figure(figsize=(48,24))
plt.plot(somalia_df['start_date2'], somalia_df['high_fatality_estimate'], marker='o', linestyle='-',color='red')
plt.title("casualties in Somalia 1990-2024",fontsize=30)
plt.xlabel("Year",fontsize=30)
plt.ylabel("Casualties", fontsize=30)
plt.grid(True)
plt.tight_layout()
plt.show()
```
