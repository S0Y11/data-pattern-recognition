```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
casualties = ['Americans killed', 'Americans Wounded','Somali Killed', 'Somali Wounded','UN personnel' ]
number_of_losses = [18,84,312,814,13]
```

```python
plt.bar(casualties,number_of_losses, color='green', alpha=0.4)
plt.title('Battle of Mogadishu 3-4 October 1993 Casualties(*Estimated)',fontsize=7)
plt.xlabel('number_of_losses', fontsize=7)
plt.ylabel('casualties', fontsize=7)
plt.show()
```

```python
categories = ['Killed', 'Wounded']
Americans = [18,84]
Somalis = [312,814]

x = np.arange(len(categories))
width = 0.1

plt.figure(figsize=(4,6))

plt.bar(x , Somalis,width,label='Somalis',color='grey', alpha=0.4)
plt.bar(x , Americans, width, label='Americans',color='red')

plt.xlabel('Type of Casualty')
plt.ylabel('Number of People')
plt.title('Battle of Mogadishu 3-4 October 1993: Casualties Comparison' )
plt.xticks(x, categories)
plt.legend()

plt.tight_layout()
plt.show()
```
