```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.patches as mpatches
```

```python
df.plot(kind="scatter", x="Longitude ", y="Latitude ", alpha=0.4 )
plt.figure(figsize=(10,8))
for i ,row in df.iterrows():
 plt.text(row['Longitude '] + 0.1, row['Latitude ']+ 0.1, row['Region'],fontsize=8)
                  
plt.title('Occupation pre-colonialism and post civil war by Region')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Occupied by')
plt.grid(True)
plt.show() 
```

```python
fig, ax = plt.subplots(figsize=(20,16))

ax.scatter(df['Longitude '], df['Latitude '], alpha=1, color='black')

for i, row in df.iterrows():
    x = row['Longitude ']
    y = row['Latitude ']
     
    ax.text(x + 0.1,y + 0.1, row['Region'], fontsize=11, ha='left')
    ax.text(x + 0.1,y - 0.1, row['occupied by'], fontsize=9,color='blue', ha='left')
    ax.text(x + 0.1,y - 0.3, row['Former colonial power'], fontsize=9,color='red', ha='left')
    
ax.set_xlabel("Longitude ")
ax.set_ylabel("Latitude ")
ax.set_title("Occupation pre-colonialism and post civil war by Region") 
plt.grid(True)

legend_labels = [
    mpatches.Patch(color='black', label='Region'),
    mpatches.Patch(color='red', label='Occuped by'),
    mpatches.Patch(color='blue', label='Former colonial power')
]
ax.legend(handles=legend_labels, loc='upper left')

plt.show()
```
